```python
#Name-Aditya Agrawal
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
data=pd.read_csv('Desktop/SampleSuperstore.csv')
data
```

```python
X=data[['Discount']]
y=data[['Profit']]
```

```python
X2=data[['Quantity']]
y2=data[['Profit']]
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
y_pred=regressor.predict(X)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X2, y2)
y_pred2=regressor.predict(X2)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(X,y)
plt.title('Discount vs Profit')  
plt.xlabel('Discount')  
plt.ylabel('Profit') 
plt.grid()
plt.show()
```

```python
#Lower discounts lead to higher profits as can be seen in the above plot
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(X2,y2)
plt.title('Quantity vs Profit')  
plt.xlabel('Quantity')  
plt.ylabel('Profit') 
plt.grid()
plt.show()
```
